```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Generate synthetic data for two variables (x and y)
n_samples = 200
x = np.random.normal(loc=50, scale=15, size=n_samples)  # Variable X
y = 2 * x + np.random.normal(loc=0, scale=10, size=n_samples)  # Variable Y, with some noise

# Create DataFrame
df = pd.DataFrame({
    'X': x,
    'Y': y
})

# Visualize the relationship between X and Y using a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='X', y='Y', data=df, color='purple')
plt.title('Scatter Plot of X vs Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

3) Create a dataset representing categories and their corresponding values. Compare different categories
based on numerical values.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Generate synthetic categorical and numerical data
n_samples = 300
categories = ['Category_A', 'Category_B', 'Category_C', 'Category_D']

# Create categorical data
category_data = np.random.choice(categories, size=n_samples)

# Generate numerical data
numerical_data = np.random.normal(loc=50, scale=15, size=n_samples)  # Normal distribution

# Add variability based on category (optional for differentiation)
numerical_data[category_data == 'Category_A'] += 10
numerical_data[category_data == 'Category_B'] -= 5
numerical_data[category_data == 'Category_C'] += 3
numerical_data[category_data == 'Category_D'] -= 3

# Create DataFrame
df = pd.DataFrame({
    'Category': category_data,
    'Value': numerical_data
})

# Visualize the distribution of the numerical variable across categories using a boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Category', y='Value', data=df, palette='Set2')
plt.title('Comparison of Numerical Values Across Categories')
plt.show()
```

5)  Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

# Generate synthetic data with correlated features
# Number of samples and features
n_samples = 100
n_features = 5

# Create correlated features using a multivariate normal distribution
mean = np.zeros(n_features)
cov = np.full((n_features, n_features), 0.8)  # High correlation
np.fill_diagonal(cov, 1)  # Set diagonal to 1 (variance)
X = np.random.multivariate_normal(mean, cov, size=n_samples)

# Convert to DataFrame for easier handling
df = pd.DataFrame(X, columns=[f'Feature_{i+1}' for i in range(n_features)])

# Compute correlation matrix
correlation_matrix = df.corr()

# Visualize the correlation matrix using seaborn heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd
import plotly.express as px

# Set the random seed for reproducibility
np.random.seed(20)

# Generate the dataset for sales data across months and days
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

# Create DataFrame
df = pd.DataFrame(data)

# Pivot the DataFrame to structure data for the heatmap
heatmap_data = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

# Create a heatmap to visualize sales data across months and days
fig = px.imshow(heatmap_data,
                labels={'x': 'Month', 'y': 'Day', 'color': 'Average Sales'},
                title='Sales Variation Across Months and Days')

# Show the plot
fig.show()
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Generate x, y, and z data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the data to fit the format for Plotly
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}

# Create DataFrame
df = pd.DataFrame(data)

# Create a 3D surface plot using Plotly
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# Add title and axis labels
fig.update_layout(
    title='3D Surface Plot of sin(sqrt(x^2 + y^2))',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)

# Show the plot
fig.show()
```

1) Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the data
x = np.linspace(0, 10, 100)  # x values from 0 to 10
y = np.sin(x)  # y values as the sine of x

# Output the plot inline in a Jupyter notebook
output_notebook()

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='Y')

# Plot the sine wave
p.line(x, y, legend_label="sin(x)", line_width=2, color="blue")

# Show the plot
show(p)
```

2) Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.palettes import Spectral11

# Generate random data for x and y
np.random.seed(42)
x = np.random.rand(100) * 10  # Random x values between 0 and 10
y = np.random.rand(100) * 10  # Random y values between 0 and 10

# Generate random sizes and colors for markers
sizes = np.random.randint(5, 50, 100)  # Random sizes for the markers
colors = np.random.choice(Spectral11, 100)  # Random colors from the Spectral11 palette

# Create a DataFrame
data = pd.DataFrame({
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': colors
})

# Output the plot inline in a Jupyter notebook
output_notebook()

# Create a Bokeh figure
p = figure(title="Scatter Plot with Varying Marker Sizes and Colors",
           x_axis_label='X', y_axis_label='Y')

# Plot the scatter plot
p.scatter(x='x', y='y', size='sizes', color='colors', legend_field='colors', source=data, fill_alpha=0.6)

# Show the plot
show(p)
```

4) Create a Bokeh histogram to visualize the distribution of the given data.

data_hist = np.random.randn(1000)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Generate random data (normally distributed)
data_hist = np.random.randn(1000)

# Compute histogram data
hist, edges = np.histogram(data_hist, bins=30)

# Output the plot inline in a Jupyter notebook
output_notebook()

# Create a Bokeh figure
p = figure(title="Histogram of Random Data",
           x_axis_label='Value', y_axis_label='Frequency',
           tools="pan,box_zoom,reset")

# Plot the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="skyblue", line_color="white", alpha=0.7)

# Show the plot
show(p)
```
